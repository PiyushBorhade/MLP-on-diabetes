```python
X = df.drop(columns=['Outcome'])
y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=11)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train_trf = sc.fit_transform(X_train)
X_test_trf = sc.transform(X_test)
```

```python
X_train_trf.shape
```

```python
X_test_trf.shape
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(8,activation='sigmoid',input_dim=8))
model.add(Dense(8,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X_train_trf,y_train,batch_size=50,epochs=100,verbose=1,validation_split=0.25)
```

```python
model.predict(X_test)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_pred = y_pred.argmax(axis=-1)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
